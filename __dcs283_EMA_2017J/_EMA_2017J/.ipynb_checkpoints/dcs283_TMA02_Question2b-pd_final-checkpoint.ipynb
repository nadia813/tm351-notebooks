# Import the KS2 data

Most of the field names are given in the `ks2_meta` file, so we'll use that to keep track of the types of various columns. 

```python
ks2cols = pd.read_csv('data/2015-2016/ks2_meta.csv')
ks2cols['Field Name'] = ks2cols['Field Name'].apply(lambda r: r.strip(),)
ks2cols
```

Some columns contain integers, but _**pandas**_ will treat any numeric column with `na` values as `float64`, due to NumPy's number type hierarchy. 

```python
int_cols = [c for c in ks2cols['Field Name'] 
            if c.startswith('T')
            if c not in ['TOWN', 'TELNUM', 'TKS1AVERAGE']]
int_cols += ['RECTYPE', 'ALPHAIND', 'LEA', 'ESTAB', 'URN', 'URN_AC', 'ICLOSE']
int_cols += ['READ_AVERAGE', 'GPS_AVERAGE', 'MAT_AVERAGE']
```

Some columns contain percentages. We'll convert these to floating point numbers on import.

Note that we also need to handle the case of `SUPP` and `NEW` in the data.

```python
def p2f(x):
    if x.strip('%').isnumeric():
        return float(x.strip('%'))/100
    elif x in ['SUPP', 'NEW', 'LOWCOV', 'NA', '']:
        return 0.0
    else:
        return x
```

These are the columns to try to convert from percentages. Note that we can be generous here, as columns like PCODE (postcode) will return the original value if the conversion fails.

```python
percent_cols = [f for f in ks2cols['Field Name'] if f.startswith('P')]
percent_cols += ['WRITCOV', 'MATCOV', 'READCOV'] 
percent_cols += ['PTMAT_HIGH', 'PTREAD_HIGH', 'PSENELSAPK', 'PSENELK', 'PTGPS_HIGH']
percent_converters = {c: p2f for c in percent_cols}
```

```python
ks2_df = pd.read_csv('data/2015-2016/england_ks2final.csv', 
                   na_values=['SUPP', 'NEW', 'LOWCOV', 'NA', ''],
                   converters=percent_converters)
```

Drop the summary rows, keeping just the rows for mainstream and special schools.

```python
ks2_df = ks2_df[(ks2_df['RECTYPE'] == 1) | (ks2_df['RECTYPE'] == 2)]
```

Convert everything to numbers, if possible.

```python
ks2_df = ks2_df.apply(pd.to_numeric, errors='ignore')
```

Merge the LEA data into the school data

```python
ks2_df = pd.merge(ks2_df, leas_df, on=['LEA'])
ks2_df.head().T
```

I want to make the plot clearer so we can identify which school is which.

```python
# Plotting Clearer version.
```

```python
# Make a plotting_df so we can add colour column to it
# sort it so the legend makes more sense
scatter_df = mainstream_gp_df.copy().sort_values(['PTMAT_EXP', 'PTREAD_EXP'], ascending=False)
scatter_df
```

```python
# to add a colour column it is easier to have a numeric index
scatter_df.reset_index(inplace=True)
```

```python
# check how many colours we need
# make a list of school types
school_types = list(scatter_df['SCHOOL_TYPE'])

len(school_types)
```

```python
# pick seven colours and assign them to a list
colours = ['red','orange','yellow','green','blue','fuchsia','purple']
```

```python
# add a colour column with a default value of black
scatter_df['COLOUR'] = 'black'
```

```python
# iterate through the DataFrame and assign a colour to each school type
for i in range(len(scatter_df)):
    scatter_df.loc[(scatter_df['SCHOOL_TYPE'] == school_types[i]), ['COLOUR']] = colours[i]

scatter_df
```

```python
# Create the coloures scatter plot
scatter_df.plot.scatter(x='PTMAT_EXP',y='PTREAD_EXP',
                       s=100,
                       c=scatter_df['COLOUR'])

# add the title and axis labels
plt.title('Percentage of students acheiving expected study level')
plt.xlabel('Maths\n(% to acheive level)')
plt.ylabel('Reading\n(% to acheive level)')

# add a legend to the scatter plot
import matplotlib.patches as mpatches
# make the legend handles
legend_handles = ([mpatches.Patch(color=colours[0], label=school_types[0])]+
                 [mpatches.Patch(color=colours[1], label=school_types[1])]+
                 [mpatches.Patch(color=colours[2], label=school_types[2])]+
                 [mpatches.Patch(color=colours[3], label=school_types[3])]+
                 [mpatches.Patch(color=colours[4], label=school_types[4])]+
                 [mpatches.Patch(color=colours[5], label=school_types[5])]+
                 [mpatches.Patch(color=colours[6], label=school_types[6])])

plt.legend(handles=legend_handles, loc=(1.05, 0.5))

```

We can now clearly see that the two best performing schools are Academy converter and Voluntary aided schools.  While the least performing school over these two measures is the sponsored academy type.  It also suggests a positive correlation between both measures.  

# Statistical testing

As there appears to be a correlation between the percentages of students acheiving the expected level in maths and in reading, I will perform a pearsons_R

## Pearson's *R*² test

```python
import scipy.stats
```

```python
scipy.stats.pearsonr(mainstream_gp_df['PTMAT_EXP'],
                    mainstream_gp_df['PTREAD_EXP'])
```

This result shows a quite strong positive correlation with a very small *p* value. Looking at the data it appears that school types that have a good student achievement level in one of the two subjects also do well in the other.  However, the sample data as it has been processed is quite small in number so it may not be an accurate test in this case.

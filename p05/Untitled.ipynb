# using maps

```python
import folium
import pandas as pd
```

```python
# default map
map_ou = folium.Map(location=[52.024696, -0.7170710], width=960, height=500)
map_ou
```

```python
# create a folder for the maps generated
!mkdir -p map-output-files
# save the html file
map_ou.save('map-output-files/ou.html')
```

```python
# prettify the map a little
map_ou2 = folium.Map(location=[52.024696, -0.710710], width=600, height=300,
                     zoom_start=14, max_zoom=16)
# max zoom limits how far in we can zoom
map_ou2
```

```python
# Adding map markers
folium.Marker([52.025638, -0.702336], popup='The East Campus of the Open University').add_to(map_ou2)

map_ou2
```

## Finding the Latitude and Longitude of locations

If we are manually building the map we can use a service like http://www.itouchmap.com/latlong.html to find the latlon.

Some datasets may already include some geodata within them.

Sometimes it will be necessary to look up the lat lon data from a postcode lookup service such as:
http://www.uk-postcodes.com/
    
    

```python
# we can also add a latlon finder to the folium map
folium.LatLngPopup().add_to(map_ou2)
map_ou2
```

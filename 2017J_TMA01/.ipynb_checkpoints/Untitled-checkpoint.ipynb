```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import re
```

```python
# read in the csv files that were cleaned in OpenRefine
# apples
apples_df = pd.read_csv('data/fruitveg_dcs.csv')
apples_df.head(5)
```

```python
# # lets summarize get the mean price for each variety
# # make a list for ease of mainipulation
month_list = ['Jan','Feb','Mar','Apr','May','Jun',
              'Jul','Aug','Sep','Oct','Nov','Dec']
# # create the new column
apples_df['Average'] = apples_df[month_list].mean(axis=1)
# # This post was very helpful
# # https://stackoverflow.com/questions/25748683/pandas-sum-dataframe-rows-for-given-columns

# drop the now unnecessary month columns
apples_df.drop(month_list, axis=1, inplace=True)

apples_df.head()
```

```python
apples_df = apples_df.groupby(['Variety','Year'],as_index=False)['Average'].mean()

apples_df.head()
```

```python
varieties = set(list(apples_df['Variety']))
varieties
```

```python
# orchards
orchards_df = pd.read_csv('data/orchards_dcs.csv')
orchards_df.head(5)
```

```python
# rename the columns of orchard to match the apples columsn so we can join the two tables
or_cols = orchards_df.columns
or_cols
```

```python
orchards_df.rename(columns={'WorcesterPearmain':'Worcester Pearmain',
                           'Cox(andclones)':'Cox',
                           'EgremontRusset': 'Egremont Russet',
                           'Gala(andclones)': 'Gala',
                           'Jonagold(andclones)': 'Jonagold',
                           'Other dessert varieties': 'Other',
                           'Fiesta/RedPippin': 'Red Pippin'}, inplace=True)
orchards_df.head()

# check for any other unmatched variety names
varieties ^ set(orchards_df.columns)
```

```python
# Note there was an inconsistency in the spelling of jonogold between the two datasets
# lets clean that in the apples_df as that was incorrectly spelled

apples_df.replace(to_replace='Jonogold', value='Jonagold', inplace=True)
apples_df.head()
```

```python
orchards_df.head()
# orchards_df.dtypes
# shows that year is an int64
orchards_df['Year'] = orchards_df['Year'].astype(str)
```

```python
orchards_df
```

```python
orchards_df = pd.melt(orchards_df, id_vars='Year',var_name='Apple variety', value_name='Orchard space')
orchards_df = orchards_df[['Apple variety','Year','Orchard space']]
```

```python
orchards_df[:15]
```

```python
orchards_df.dtypes
```

```python
apples_df.columns = ['Apple variety','Year','Average wholesale price']
```

```python
# I was struggling to get the two datasets to merge.  Then I thought, I haven't double checked the data types of both datasets

# apples_df.dtypes
# shows that year is an int64
apples_df['Year'] = apples_df['Year'].astype(str)
```

```python
# join merge the two datasets
merged_df = pd.merge(apples_df, orchards_df, on=['Apple variety','Year'], how='inner')
merged_df.head()
```

```python
# There are some missing values that we should drop
merged_df.dropna(inplace=True)
```

```python
# and it is a good time to cleand the values of data
merged_df['Average wholesale price'] = round(merged_df['Average wholesale price']*100,1)

merged_df.head()
```

```python
merged_df.describe()
```

```python
merged_df.groupby('Year').mean().plot(y='Orchard space')
plt.xlabel('Year')
plt.ylabel('Tree area')
plt.title('Average dessert apple orchard space')
```

```python
# Now let's look at the average price of all apples over the same period
merged_df.groupby('Year').mean().plot(y='Average wholesale price')
plt.xlabel('Year')
plt.ylabel('Price in Pence')
plt.title('Average wholesale price of all dessert apples')
```

```python
orch_plot = merged_df[['Year', 'Orchard space']].groupby('Year').mean()
orch_plot.head()

price_plot = merged_df[['Year','Average wholesale price']].groupby('Year').mean()
# price_plot.head()
```

```python
fig = plt.figure()
# set up axes for first subplot
orchardAxes = fig.add_subplot(111)

orch_plot.plot(ax=orchardAxes, color='green', ylim=[550,850] )
orchardAxes.set_ylabel('Total orchard space (tree area)')

# set up the second plot sharing the same x axes
priceAxes = orchardAxes.twinx()

price_plot.plot(ax=priceAxes, ylim=[40,75],color='blue')
priceAxes.set_ylabel('Wholesale price (pence/Kg)')

# Arrange the Axes Legends
orchardAxes.legend(loc=2, frameon=False, fontsize=10)
priceAxes.legend(loc=4, frameon=False, fontsize=10)

# Add the title
plt.title('Eight year price increase of apple prices\nand decrease in orchard space')
```

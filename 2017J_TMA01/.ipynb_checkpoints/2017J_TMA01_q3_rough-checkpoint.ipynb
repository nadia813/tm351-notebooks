```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sn
```

```python
fruitveg_df=pd.read_csv('datasets/fruitveg.csv', sep='\t')
fruitveg_df.iloc[15:45]
```

```python
orchard_df=pd.read_csv('datasets/orchfruit_ap&pr_30may13.csv')
orchard_df
```

```python
apples2004_df=fruitveg_df[['Unnamed: 1', 'Quality', 'Jan', 'Feb', 'Mar', 'Apr', 'May', \
                           'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].iloc[15:45]

apples2004_df
```

```python
apples2004_df=fruitveg_df[['Unnamed: 1', 'Jan', 'Feb', 'Mar', 'Apr', 'May', \
                           'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']].iloc[[17, 20, 23, 26, 29, 32, 35, 41, 44]]

apples2004_df
```

```python
for col in ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']:
    apples2004_df[col]=pd.to_numeric(apples2004_df[col])
    
apples2004_df
```

Now, I want the average per month of each apple. I won't bother with the 'other' columns.

```python
ap_df=apples2004_df.rename({i:apples2004_df['Unnamed: 1'][i] for i in apples2004_df.index}).drop('Unnamed: 1', axis=1)
ap_df=ap_df.T

ap_df
```

```python
orchard_df.loc[5]
```

```python
acreage_ss=pd.Series()

acreage_ss['Cox’s Orange-group']=3144
acreage_ss['Discovery']=301
acreage_ss['Egremont Russet']=308
acreage_ss['Spartan']=164
acreage_ss['Worcester Pearmain']=213
acreage_ss['Red Pippin']=np.nan
acreage_ss['Gala']=669
acreage_ss['Jonogold – group']=231
acreage_ss['Katy']=np.nan

acreage_ss
```

Now, need to plot each type of apple against the acreage.

```python
colours=['red', 'blue', 'green', 'yellow', 'purple', 'white', 'black', 'darkgreen', 'darkblue']


l=list(ap_df.columns)
l.remove('acreage')
l

a=l[0]

ap_df['acreage']=acreage_ss[a]
ax=ap_df.plot.scatter('acreage', a, c=colours[0], label=a)

for (i, a) in enumerate(l[1:]):
    ap_df['acreage']=acreage_ss[a]
    ap_df.plot.scatter('acreage', a, label=a, c=colours[i+1], ax=ax)

plt.xlim(0, 1000)
plt.ylim(0, 1)
plt.legend(loc=4)
plt.ylabel('Units / £/kg')
```

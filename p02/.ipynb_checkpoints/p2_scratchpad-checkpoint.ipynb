### KML - mapping locations

```python
from fastkml import kml
k = kml.KML()

# open bytestream of carpark data
doc = open('data/CarParks.kml','rb').read()
k.from_string(doc)

# parse the locations of carparks from the file
locations = dict()
for feature in k.features():
    for placemark in feature.features():
        locations.update({placemark.name: (placemark.geometry.y, placemark.geometry.x)})
list(locations)
```

```python
# now lets quickly map the markers - req internet
import folium # we will look at more in p5

carparks = folium.Map(location=[50.68, -1.2667], width=960, height=500, zoom_start=11)

for name, location in locations.items():
    folium.CircleMarker(location=location,
                       popup=name,
                       radius = 20,
                       fill_color='green',
                       fill_opacity=0.2
                       ).add_to(carparks)
# display the map
carparks
```

```python
# create an html file for the map
carparks.save('data/snowkuma_carparks.html')
```

### YAML
No direct support but can use _PyYaml_ to load in file and convert to pydict then DF.

WARNING use safe_load() safe_load_all() to avoid executables

http://pyyaml.org/wiki/PyYAMLDocumentation.

```python
import yaml
# yaml.load() will accept a single document string, and parse it to generate
# a Python dict, so will yaml.safe_load():
document = """
image:
    width: 800
    height: 600
    title:  View from 15th Floor
    thumbnail:
        url: http://www.example.com/image/481989943
        height: 125
        width:  100
        animated : false
    IDs:
        - 116
        - 943
        - 234
        - 38793
"""
parsedYAML = yaml.safe_load(document)
parsedYAML
```

```python
# Here 'document.yaml' contains a single YAML document.
stream = open('data/document.yaml', 'r') 
yaml.safe_load(stream)
```

```python
# We can also cast a dict to YAML using the yaml.dump() function applied to a dict:
print(yaml.dump(parsedYAML))
```

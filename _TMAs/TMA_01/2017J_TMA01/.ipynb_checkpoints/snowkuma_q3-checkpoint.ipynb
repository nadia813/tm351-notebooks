```python
fruitveg_df = pd.read_csv('data/fruitveg.csv', sep='\t')
fruitveg_df.rename(columns={'2004': 'Fruit','Unnamed: 1': 'Variety'}, inplace=True)
fruitveg_df[:5]
```

```python
fruitveg_df['Year'] = fruitveg_df['Fruit']
fruitveg_df.columns = ['Year','Fruit','Variety','Quality','Units','Jan','Feb','Mar','Apr','']
```

```python
fruitveg_df['Fruit'].fillna(method='ffill', inplace=True)
apples_df = fruitveg_df.loc[fruitveg_df['Fruit'] == 'Dessert Apples']
apples_df[:5]
```
